# Second-week Results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from mpl_toolkits import mplot3d
from matplotlib.animation import FuncAnimation
from pandas.core.frame import DataFrame
```

```python
data = pd.read_csv('Data.csv', sep = ",")

data_index = data.set_index(['UID','No']).sort_index()
```

```python
# Plot x data for each sensor (UID)
```

```python
def data_plot_x(num):
    data_index_x = data_index.loc[num]
    data_x = pd.pivot_table(data_index_x, ['X'], 'No')
    data_x.plot()
    plt.show()
```

```python
%matplotlib inline
num = 1000101017
data_plot_x(num)
```

```python
# Plot x data for all sensors together
```

# For each sensor plotting changing x, y, z

```python
%matplotlib inline
num = 1000101017
fig, axs = plt.subplots(10, sharex=True, sharey=True, figsize=(16,50))
plt.ylim(-10, 95)

i = 0
for i in range(10):
    ax = axs[i]
    data_per_UID = data_index.loc[num]
    data_per_UID_table = pd.pivot_table(data_per_UID, ['X', 'Y', 'Z', 'T'], 'No')
    data_per_UID_table.plot(ax=ax)
    num = num + 1
    i = i + 1
```

# Plot X, Y, Z vs T for each sensor

```python
%matplotlib inline
num = 1000101017

i = 0
for i in range(10):
    data_per_UID_temp = data_index.loc[num]
    data_per_UID_table_temp = pd.pivot_table(data_per_UID_temp, ['X', 'Y', 'Z'], 'T')
    data_per_UID_table_temp.plot()
    num = num + 1
    i = i + 1
```

# Plot angle changed for each sensor

```python
df = pd.read_csv('data.csv',delimiter=',')
index = df.index
columns = df.columns
values = df.values
```

```python
df2 = df
df2 = df2.drop(['No'], axis=1)
df17 = df2.loc[df2['UID'] == 1000101017]
df18 = df2.loc[df2['UID'] == 1000101018]
df19 = df2.loc[df2['UID'] == 1000101019]
df20 = df2.loc[df2['UID'] == 1000101020]
df21 = df2.loc[df2['UID'] == 1000101021]
df22 = df2.loc[df2['UID'] == 1000101022]
df23 = df2.loc[df2['UID'] == 1000101023]
df24 = df2.loc[df2['UID'] == 1000101024]
df25 = df2.loc[df2['UID'] == 1000101025]
df26 = df2.loc[df2['UID'] == 1000101026]
#ax.quiver(0, 0, 0, df18['X'].values, df18['Y'].values, df18['Z'].values, length=0.1, normalize=True)
```

```python
df17_table = pd.pivot_table(df17, ['X', 'Y', 'Z'], df17.index)
df17_angle = df17_table.values

angleArray_17,index = [],[]
for i,j in enumerate(df17_angle[0:len(df17_angle)-1]):
    a = df17_angle[i]
    b = df17_angle[i+1]
    A = np.linalg.norm(a)
    B = np.linalg.norm(b)
    ans = np.arccos((np.dot(a,b))/(A*B))
    angleArray_17.append(ans)
    index.append(i)

A17_dic = {'angleArray_17': angleArray_17}
A17 = DataFrame(A17_dic)

No_17 = A17.index
#A17_dic = {'No_17': No_17, 'angle_changed_17': angleArray_17}
#A17 = DataFrame(A17_dic)
```
