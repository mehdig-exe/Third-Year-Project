```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import plotly.io as pio
import matplotlib, random
import math
import ipywidgets as widgets
from IPython.display import display,clear_output


url1 = 'https://raw.githubusercontent.com/mehdighemati/Third-Year-Project/master/Mehdi/Parse/OTN%20Data.txt'
url2 = 'https://raw.githubusercontent.com/mehdighemati/Third-Year-Project/master/Mehdi/Parse/hub.txt'
url3 = 'https://raw.githubusercontent.com/mehdighemati/Third-Year-Project/master/Mehdi/Parse/OTN-StabilityTestLog.txt'

df1 = pd.read_csv(url1, error_bad_lines=False)
df2 = pd.read_csv(url2, names=list(["PCTime","UUID","Vbatt","No","year","month","day","hour","minute","second","X","Y","Z","Temp"]))
df3 = pd.read_csv(url3, names=list(["PCTime","UUID","Vbatt","No","year","month","day","hour","minute","second","X","Y","Z","Temp"]))
df2 = df2.astype(object) #When shifting to the right by 1 to a column with a different type it makes it NaN - this prevents it for now
df2 = df2.drop(df2[df2.count(axis=1) < 8].index)
check = df2.PCTime.str.contains('\d\d(\/)\d\d(\/)\d\d\d\d')
updated = df2[~check].shift(periods=1, axis='columns')
updated = updated.fillna("Change")
df2.update(updated)
df2 = df2.replace({'PCTime': {'Change': np.nan}})
df2
df1 = df1.rename(columns={"UID":"UUID","V":"Vbatt","T":"Temp"})

def createDateTime(dfDT):
  test = pd.to_datetime(dfDT['year'].astype(int).astype(str)+'/'+dfDT['month'].astype(int).astype(str)+'/'+ dfDT['day'].astype(int).astype(str)+' '+dfDT['hour'].astype(int).astype(str)+':'+dfDT['minute'].astype(int).astype(str)+':'+dfDT['second'].astype(int).astype(str), format = '%y/%m/%d %H:%M:%S').dt.strftime('%y/%m/%d %H:%M:%S')
  test = pd.DataFrame(test)
  test.rename(columns={0:'DateTime'}, inplace=True)
  dfDT['DateTime'] = pd.Series(test['DateTime'])

createDateTime(df2)
createDateTime(df3)

#df1 check
df1.UUID = df1.UUID.apply(str)
check2 = df1.UUID.str.len() ==4
cleandf1 = df1[~check2]

df2 = df2.append(cleandf1, ignore_index=True)
df2 = df2.append(df3, ignore_index=True)

dfs = dict(tuple(df2.groupby('UUID')))
#dfsl is the (d)ata(f)rame of unique (s)ensors within a (l)ist 
dfsl = list(tuple(df2.groupby('UUID')))

#To check the unique Sensor UUIDs
choices = []
tracers = []
for i,j in enumerate(dfsl):
  choices.append("Sensor " + str(j[0]) + " X")
  choices.append("Sensor " + str(j[0]) + " Y")
  choices.append("Sensor " + str(j[0]) + " Z")
  choices.append("Sensor " + str(j[0]) + " Temp")
  choices.append("Sensor " + str(j[0]) + " Vbatt")
  tracers.append("X")
  tracers.append("Y")
  tracers.append("Z")
  tracers.append("Temp")
  tracers.append("Vbatt")
choices

hex_colors_dic = {}
rgb_colors_dic = {}
hex_colors_only = []
for name, hex in matplotlib.colors.cnames.items():
    hex_colors_only.append(hex)
    hex_colors_dic[name] = hex
    rgb_colors_dic[name] = matplotlib.colors.to_rgb(hex)

```

```python
sensorSelect = widgets.SelectMultiple(
    options=choices,
    value=[choices[0]],
    #rows=10,
    description='Sensors',
    disabled=False
)
display(sensorSelect)
button = widgets.Button(description="Generate")
output = widgets.Output()

display(button, output)

#Main figure cell
def graphFigure(order,selected):
  fig = go.Figure()

  buttonslist =[]
  labelslist = []

  def getvisiblelist(num):
    visiblecheck = []
    for i,j in enumerate(order):
      if i != num:
        visiblecheck.append(False)
      else:
        visiblecheck.append(True)
    return visiblecheck

  def getlabel(num):
    element = (str(selected[num]))
    return(element)

  for i,j in enumerate(order):
    element = dict(label=getlabel(i),
                  method="update",
                  args=[{"visible": getvisiblelist(i)},
                        {"title": str(selected[i])}])
    buttonslist.append(element)

  for i,j in enumerate(order):
    fig.add_trace(
        go.Scatter(x=list(dfsl[math.floor(j/5)][1].DateTime),
                  y=list(dfsl[math.floor(j/5)][1][tracers[j]]),
                  name=str(selected[i]),
                  line=dict(color=random.choice(hex_colors_only))))

  fig.update_layout(
      updatemenus=[
          dict(
              active=0,
              buttons=buttonslist,
          )
      ])



  # Set title
  fig.update_layout(title_text="Graph")

  fig.show()

def graph(self):
  selected = list(sensorSelect.value)
  order = []
  for i,j in enumerate(selected):
    order.append(choices.index(j))
  order
  graphFigure(order,selected)

button.on_click(graph)
```
