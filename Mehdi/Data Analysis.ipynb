```python
crossArray = []
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection='3d')
for i,j in enumerate(x1[0:len(x1)-1]):
    crossArray.append(np.cross(x1[i],x1[i+1],))
    ax.quiver(0, 0, 0, crossArray[i][0], crossArray[i][1], crossArray[i][2], length=0.1, normalize=True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Change in Normal Vectors')


    
```

Angles between the Normals

```python

angleArray,index = [],[]

for i,j in enumerate(crossArray[0:len(crossArray)-1]):
    a = crossArray[i]
    b = crossArray[i+1]
    A = np.linalg.norm(a)
    B = np.linalg.norm(b)
    ans = np.arccos((np.dot(a,b))/(A*B))
    angleArray.append(ans)
    index.append(i)






   
```
